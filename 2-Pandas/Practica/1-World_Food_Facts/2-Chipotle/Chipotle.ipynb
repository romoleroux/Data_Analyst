### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
address = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(address, sep='\t')
chipo
```

```python
chipo.info()
```

```python
chipo['item_price'].str[1:].astype(float)
```

```python
chipo['item_price'] = chipo['item_price'].str[1:].astype(float)
chipo.info()
```

# Name of the max valued product

```python
chipo['item_price_unit'] = chipo['item_price'] / chipo['quantity']
```

```python
chipo.head()
```

```python
chipo['item_price_unit'].max()
```

```python
df_max_value = chipo[chipo['item_price_unit'] == chipo['item_price_unit'].max()]
df_max_value
```

```python
df_max_value['item_name'].unique()
```
